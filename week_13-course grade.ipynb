```python
import random
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


attendance = []
project_scores = []
project_proposal=[]
assignment_scores = []
quiz_scores=[]
res_score=[]
Total_assignment=[]
```

Score evaluators

```python
attendance_eval = 15 / 10
project_eval = 300 / 60
assignment_eval = 1000 / 30
```

```python
attendance=['A','A','A','A','A','A','A','A','A','A','A','A','A','A','A']
```

```python
project_scores=[100,100,100]
```

```python
assignment_scores=[100,95,100,100,100,100,100,82,100,100,100,100,75,80,100,75,100,85]
#considered 40 in quiz as 100 
quiz_scores=[100,100]
Total_assignment=assignment_scores+quiz_scores
```

```python
print ("Attendance:\n{}\n".format(attendance))
print ("res_score:\n{}\n".format(project_scores))
print ("Assignment Scores:\n{}\n".format(Total_assignment))
```

Evaluate attendance scores

```python
attendance_scores = []
for i in attendance:
    if i == "A":
        attendance_scores.append(1)
    else:
        attendance_scores.append(0)
```

```python
attendance_scores
```

Evaluate total obtained scores and creating 100 outcomes.

```python
list1=[]
for j in range(100):
    random_assignments = []
    for i in range(10):
        random_assignments.append(random.choice(Total_assignment))
        obt_asgnmnt_score = sum(random_assignments) / assignment_eval
        obt_prjct_score = sum(project_scores) / project_eval
        obt_attndnce_score = sum(attendance_scores) / attendance_eval
        #print(obt_asgnmnt_score)
        total_score = obt_asgnmnt_score + obt_prjct_score + obt_attndnce_score
        total_score1=round(total_score)
        list1.append(total_score1)
```
