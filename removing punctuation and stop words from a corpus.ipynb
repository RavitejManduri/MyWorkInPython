Reading each file and storing each file in separate list i.e only text files

```python
import string
import glob
i=0
docwords=[]
file_list=[]
file=[[] for x in range(15)]
path = 'week_10_txt_and_docx'

files = [f for f in glob.glob(path + "**/*.txt", recursive=True)]

for fi in files:
    #file_list.append(fi)
    print(fi)
   

    with open(fi,'r', encoding="utf8") as f:        
        for line in f:
            line=line.lower()
            for j in string.punctuation:
                line=line.replace(j,"")

            for word in line.split():
                file[i].append(word)
    i=i+1

```

```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele+","  
    
    # return string   
    return str1  
```

```python
datfile=""
i=0
for fi in files:    
    datfile+='"'+ fi +'":'
    datfile+=listToString(list(set(file[i])))+'\n'
    i=i+1
    
datfile+='"'+'week_10_txt_and_docx/week_10_document1.docx' +'":'
datfile+=listToString(list(set(file[i])))+'\n'
i=i+1
datfile+='"'+'week_10_txt_and_docx/week_10_document2.docx' +'":'
datfile+=listToString(list(set(file[i])))+'\n'
i=i+1
    
print(datfile)
```

```python
file = open("removing_punctuation.dat", "w",errors='ignore')

file.write(datfile)
file.close()
```
