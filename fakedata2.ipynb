I stored the randomly selected names in "n_list.txt" file.

```python
import random
import numpy
import csv
from random import randint
import os
from os import path
if(path.exists("fakedata.csv")):
    os.remove("fakedata.csv")
nameslist=open("n_list.txt","r")
n_list=nameslist.read()
```

For creating mail ID's,I stored few domain names.

```python
n_list=n_list.split()
domain_list=["gmail.com","twitter.com","ben.edu","umbc.edu","google.co.in"]
```

Below is to check whether the house is owned or rental

```python
housestatus=["own","rent"]
```

To generate the address,this is the link I referred for the states information and stored as USA states.txt.http://www.softschools.com/social_studies/state_abbreviations/

Here, I defined the address function in which I took zipcode as zipcd,flat number,country and state.

```python
def address():
    zipcd=str(randint(10000,99999))
    states=open("USA states.txt","r")
    states_list=states.read().split("\n")
    flatnum=str(randint(100,500))
    address1="flat num"+" "+str(flatnum)+" "+str(random.choice(states_list))+" "+"USA"+zipcd
    return(address1)
```

Here I took random function to generate random information and referred this code.https://www.geeksforgeeks.org/random-numbers-in-python/
I choosed name variable in which two names are selected from the n_list and stored.
To generate email ID, I choosed replace function to remove space between two names from domain_list and replaced with '@'
I stored numbers between(1000000000,9999999999) using randint function in the variable phonenumber.
Choosed years randomly from (1950,2000) and stored in variable year.
choosing randomly from the list housestatus which I stored previously.
choosing randomly a number in between (100000-1000000) using the randint function to generate annual income.
Randomly choosing a number in between (0-10) using randint function.
Here I am storing all the above in a list i.e.,completedata.



```python
completedata=[]
for i in range(100):
    name=random.choice(n_list)+" "+random.choice(n_list)
    email=str(name.replace(" "," "))+"@"+random.choice(domain_list)
    phonenumber=str(randint(1000000000,9999999999))
    year=str(randint(1950,2010))
    kids=str(randint(0,4))
    h_status=random.choice(housestatus)
    dinfo=[]
    dinfo.append(' '.join(email))
    dinfo.append(phonenumber)
    dinfo.append(address())
    dinfo.append(name)
    dinfo.append(year)
    dinfo.append(kids)
    dinfo.append(h_status)
    dinfo.append(randint(100000,1000000))
    dinfo.append(randint(0,10))
    completedata.append(dinfo)
```

```python
import pandas as pd
df=pd.DataFrame(completedata)
df
```
