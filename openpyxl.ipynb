You get an Excel spreadsheet that has two worksheets, "patient info" and "records"​

Each worksheet has data and visualizations and formulas
Each row in each worksheet represents information associated with a single person.. ​
Worksheet "patient info" has a "patient id" column; ​
Worksheet "records" has a "p_id" column​
The "patient id" values in "patient info" tab are the same variable as "p_id" in "records". Some people have information on both sheets, while other people have information only on one sheet.
Use openpyxl to copy information about patients from "records" to "patient info"​. Submit a notebook that reads the Excel spreadsheet and produces a separate spreadsheet with the following modifications:​

For each patient identified by "p_id" listed in the "records" sheet that don't exist in "patient info," create a new row in "patient info". The row from "records" should be copied to the row in "patient info."
For each patient listed in the "records" sheet where "p_id" appears "patient info" under "patient id," copy the data in the row from records to "patient info" sheet
Make no changes to the visualizations that exist in each worksheet​
Make no changes to the data in the "records"​ sheet
Write your changes to a new .xlsx file (don't overwrite the original)​ ​
Observations:​

"patient info" worksheet will have new columns (because those new columns exist in the "records" worksheet)​
"patient info" worksheet will have new rows (one new row per patient)​
There will be empty cells in "patient info" worksheet​
Use a programmatic (rather than manual) approach to identify which patients appear in both worksheets​
Some cells in both worksheets contain formulas. Copy only values from "records" to "patient info"​

```python
import openpyxl
```

open the excel sheet

```python
worksheet = openpyxl.load_workbook('week_05_homework_XLSX_openpyxl.xlsx')
```

```python
patient_info=worksheet['patient info']
```

```python
records=worksheet['records']
```

Referred this link https://stackoverflow.com/questions/49959503/in-openpyxl-how-to-find-the-max-column-number-in-a-particular-row 

```python
patientinfo_maxrows = patient_info.max_row + 1
patientinfo_maxcolumns= patient_info.max_column + 1
records_maxrows = records.max_row + 1
records_maxcolumns = records.max_column + 1
```

Referred the link https://stackoverflow.com/questions/37182528/how-to-append-data-using-openpyxl-python-to-excel-file-from-a-specified-row. Below I used append function to add the patient_info into records sheet

```python
patientinfo_patientid = []
for i in range(1,patientinfo_maxrows):
   patientinfo_patientid.append(patient_info.cell(i,1).value)
```

```python
records_p_id = []
for i in range(1,records_maxrows):
   records_p_id.append(records.cell(i,1).value)
```

```python
only_records = []
ins= list(set(patientinfo_patientid) & set(records_p_id))
only_records = [x for x in records_p_id if x not in ins]
only_records
```

Here in the  below code,I to filled the patient details in patient info from records sheet. Here we are finding row(patientinfo) of a particular patient id(of patient info) in records sheet and copying those rows(patient info) details from records to patient info sheet

```python
for r in range(1, patientinfo_maxrows):
    if patient_info.cell(r,1).value in records_p_id :
        for c in range(1,patientinfo_maxcolumns):
            patient_info.cell(r,records_maxcolumns+c).value=records.cell(r,c).value

for c in range(1,records_maxcolumns):
   patient_info.cell(1,patientinfo_maxcolumns+(c-1)).value=records.cell(1,c).value

for r in range(1,records_maxrows):
   if records.cell(r,1).value in only_records:
    row=patient_info.max_row;
    patient_info.cell(row,1).value = records.cell(r,1).value
    for c in range(1,records_maxcolumns):
        patient_info.cell(row+1,records_maxcolumns+c).value = records.cell(r,c).value
```

Here I am saving a new slsx file as 'Finalsheet'

```python
worksheet.save('finalsheet.xlsx')
worksheet.close()
```
