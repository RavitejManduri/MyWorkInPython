```python
import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup
```

```python
tree = ET.parse('xml_containing_html.xml')
root = tree.getroot()
```

Here,we are extracting the html from xml file

```python
tree1 = ET.ElementTree(file='xml_containing_html.xml')
root1=tree1.getroot()
```

Here,we take a child attribuite to get the Unique id i.e; author and get printed all the authors in the xml file

```python
for child in root1:
    print (child.get('author'))
```

Here,by using the for loop we extract the xml content and we used Beautifulsoup() to parse the html 

The BeautifulSoup object contains a method named find_all() which allows us to deserialize that structure as a list of tags

len() is used to count the no.of tags are present.

```python
for child in root1:
    print (child.get('author'))
    for attr in child:
        if(attr.tag=='content'):
            #print(attr.text)       #here we can print the html content.
            my_text=attr.text
            soup = BeautifulSoup(my_text)
            tags = soup.find_all('a')
            print(type(tags))
            print(len(tags))
```
