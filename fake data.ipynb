# Address:
This is a function which return a complete random address which has flat no,state,Country,zipcode.

```python
def address():
    zipcode=str(randint(10000,99999))
    states=open("states.txt","r")
    states_list=states.read().split("\n")
    
    flat=str(randint(101,501))
    address1="apt"+" "+str(flat)+" "+str(random.choice(states_list))+" "+"USA "+zipcode
    return(address1)
    
```

To generate the data randomly I used the code(random function) from the below link

# Name:
I am choosing two names from the names list and joining them and storing in name variable.
# Email:
In this from the name I choose previosly I am remmoving space in the middle and appending '@' and a random domain from the domain_list variable.
# Number:
number is the variable where I am storing the phone numbers of length 10.
# YEAR:
choosing years randomly from the range 1960-1999 and storing in variable year.
# Home status(Rent/own)
choosing randomly from the list status which I stored previously.
# Annual Income:
Randomly choosing a number in between (50000-200000) using the function randint.
# No of speeding Tickets:
Randomly choosing a number in between (0-20) using the function randint.
# Finally:
In the end I am storing all the above information into a list 'l'.Its basically a list of lists.

```python
l = []
for i in range(200):
    #NAME:
    name=random.choice(name_list)+" "+random.choice(name_list)
    
    #EMAIL:
    email=str(name.replace(" ",""))+"@"+random.choice(domain_list)
    
    #NUMBER:
    number=""
    for j in range(10):
        if(j!=0):
            number=number+str(randint(0, 9))
        else:
            number=number+str(randint(1, 9))
    
    #YEAR:
    year=str(randint(1960,1999))
    
    #NO of Kids:
    kids=str(randint(0,3))
    
    #House Status
    home_status=random.choice(status)
    
    data_list=[]
    data_list.append(''.join(email))
    data_list.append(number)
    data_list.append(address())
    data_list.append(name)
    data_list.append(year)
    data_list.append(kids)
    data_list.append(home_status)
    data_list.append(randint(50000,200000))
    data_list.append(randint(0,20))
    print(data_list)
    l.append(data_list)

    
   
    
    
    
    
```
